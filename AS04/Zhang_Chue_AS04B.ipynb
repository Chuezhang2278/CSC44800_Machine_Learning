```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import graphviz
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz
from sklearn import tree
from sklearn.metrics import classification_report

iris = datasets.load_iris()
X = iris.data
y = iris.target

# SIDE NOTES
# Code template taken from professor, AS03 hw 
# My work may or may not be correct, please double check before using
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.50)
```

# Entropy, Depth = 5

```python
tree = DecisionTreeClassifier(criterion='entropy',max_depth=5, random_state=0)
tree.fit(X_train, y_train)

print('depth 5:', tree.score(X_test, y_test))

export_graphviz(tree, out_file='T1.dot', feature_names=['sepal length','sepal width','petal length','petal width'])
with open('T1.dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

# Entropy, Depth = 10

```python
tree = DecisionTreeClassifier(criterion='entropy',max_depth=10, random_state=1)
tree.fit(X_train, y_train)

print('depth 10:', tree.score(X_test, y_test))

export_graphviz(tree, out_file='T2.dot', feature_names=['sepal length','sepal width','petal length','petal width'])
with open('T2.dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

# Gini, Depth = 5

```python
tree = DecisionTreeClassifier(criterion='gini',max_depth=5, random_state=1)
tree.fit(X_train, y_train)

print('depth :', tree.score(X_test, y_test))

export_graphviz(tree, out_file='T3.dot', feature_names=['sepal length','sepal width','petal length','petal width'])
with open('T3.dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

# Gini, Depth = 10

```python
tree = DecisionTreeClassifier(criterion='gini',max_depth=10, random_state=1)
tree.fit(X_train, y_train)

print('depth 10:', tree.score(X_test, y_test))

export_graphviz(tree, out_file='T4.dot', feature_names=['sepal length','sepal width','petal length','petal width'])
with open('T4.dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```
